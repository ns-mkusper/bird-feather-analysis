# 🚀 YOLOv12-seg Fine-Tuning for Feather Isolation
This notebook leverages the massive compute cluster unified memory to fine-tune the SOTA **YOLOv12 Attention-Centric Segmentation Model** on our SAM2-annotated ground truth dataset.

```python
from ultralytics import YOLO

# Initialize the SOTA YOLOv12 Segmentation model
# (Ultralytics supports YOLOv12 architecture parameters implicitly via the v12 weights)
model = YOLO('yolov12n-seg.pt')

print("Starting Distributed MPS Training Loop...")
# We explicitly use 'mps' device, 100 epochs, and a batch size optimized for unified memory
results = model.train(
    data='../data/autolabel/dataset.yaml', # Ensure this points to the generated Roboflow export
    epochs=100,
    batch=16,
    device='mps',
    project='../models',
    name='yolov12_feather_seg',
    plots=True
)

print("Training Complete! The best weights are saved to ../models/yolov12_feather_seg/weights/best.pt")
```
